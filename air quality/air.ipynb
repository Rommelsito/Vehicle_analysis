#### Tratamiento del Dataset
* Tratamiento de la columna [Message]

```python
df_1 = df_air.copy()
df_1.head()
```

```python
df_1_drop = df_1.drop(columns = "Message")
df_1_drop.head()
```

#### Tratamiento de la columna [Start_Date]

```python
df_2 = df_1_drop.copy()
df_2.head()
```

```python
df_2["Start_Date"] = pd.to_datetime(df_2["Start_Date"], format = "%m/%d/%Y")
df_2.head()
```

```python
#Modificando el nombre en la columna
df_2 = df_2.rename(columns = {"Start_Date":"Start Date"})
df_2.head()
```

```python
df_2.info()
```

#### Trabajando con la columna Start Date
