```python
import pandas as pd
import numpy as np
```

```python
df_store_sales = pd.read_csv('sales_store.csv')
df_store_sales.head()
```

```python
df_store_sales.describe()
```

```python
df_store_sales.info()
```

#### Tratamiento de la columna Date

```python
df_1 = df_store_sales.copy()
df_1.head()
```

```python
df_1['Date'] = pd.to_datetime(df_1['Date'], format ='%d/%m/%Y')
df_1.info()
```

```python
#Creando columna por año 
df_1['Year'] = pd.DatetimeIndex(df_1['Date']).year
df_1.head()
```

```python
df_1['IsHoliday'].replace({False: 0, True: 1}, inplace = True)
df_1
```

```python
df_1.reset_index(inplace = True)
df_1.head()
```

#### Categorizando por años

```python
df_2 = df_1.copy()
df_2.head()
```

```python
df_2['Year'].unique()
```

```python
df_2_2010 = df_2[df_2['Year'] == 2010]
df_2_2010
```

```python
df_2_2011 = df_2[df_2['Year'] == 2011]
df_2_2011
```

```python
df_2_2012 = df_2[df_2['Year'] == 2012]
df_2_2012
```

##### Pasando a archivo CSV

```python
df_2_2010.to_csv('Data_2010.csv', index= False)
df_2_2011.to_csv('Data_2011.csv', index= False)
df_2_2012.to_csv('Data_2012.csv', index= False)
```
