```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_death = pd.read_csv('life-expectancy.csv')
df_death
```

```python
df_death.info()
```

### Eda

```python
df1 = df_death.copy()
df1.head()
```

```python
df1['Entity'].unique()
```

```python
#Renombrando columna
df1.rename(columns={'Period life expectancy at birth - Sex: all - Age: 0': 'Age'}, inplace= True)
df1.head()
```

```python
df1.tail()
```

```python
#Eliminando columna
df_dropep = df1.drop(columns = ['Code'])
df_dropep
```

```python
df2 = df_dropep.copy()
df2
```

```python
# Redondeo de los valores en edad a 2 decimales
df2.round(2)
```

```python
#Pasando datos demimales a enteros en ['Age']
df2['Age'] = df2['Age'].astype(int)
df2
```

```python
df2['Year'].describe().reset_index()
```

#### Funcion para categorizar las edades

```python
df3 = df2.copy()
df3
```

```python
df3['Year'].describe().reset_index()
```

#### Determinar esperanza de vida en:
* Afganistan

```python
df_Afghanos = df2[df2['Entity'] == 'Afghanistan']
df_Afghanos
```

```python
df_Afghanos.groupby('Entity')['Age'].describe()
```

```python
import matplotlib.pyplot as plt

# Crear la grÃ¡fica de lÃ­neas
plt.figure(figsize=(10, 6))
plt.plot(df_Afghanos['Year'], df_Afghanos['Age'], marker='o', linestyle='-', color='b')

# AÃ±adir tÃ­tulos y etiquetas
plt.title('Esperanza de vida en Afghanistan a lo largo del tiempo')
plt.xlabel('AÃ±o')
plt.ylabel('Esperanza de vida al nacer (aÃ±os)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
## Configuramos la figura
plt.figure(figsize = (10,6))
# Aplicamos un tema de seaborn
sns.set_theme(style="whitegrid")
#Graficamos los datos de Russia y Asia
plt.plot(df_Russia['Year'], df_Russia['Age'],marker = 'o', linestyle = '-', color = 'b')
plt.plot(df_Asia['Year'], df_Asia['Age'],marker = 'o', linestyle = '-', color = 'r')
# Etiquetamos los ejes y el tÃ­tulo con fuentes personalizadas
plt.title('Expentacy life in Asia and Russia', fontsize = 14, fontweight = 'bold')
plt.xlabel('Year', fontsize = 14, fontweight = 'bold')
plt.ylabel('Age',fontsize = 14, fontweight = 'bold')
plt.annotate('Punto importante', xy=(1950, 60), xytext=(1960, 70), arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
